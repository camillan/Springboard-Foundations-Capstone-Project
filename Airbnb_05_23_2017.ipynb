```python
# seaborn - for linear regression

import seaborn as sns

for column in ratings_trimmed_sf:
    if column != 'mean_review_scores':
        sns.lmplot(x=column, y='mean_review_scores', data=ratings_trimmed_sf)
        sns.plt.show()
        sns.plt.clf()
```
